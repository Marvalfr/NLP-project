```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.tree import plot_tree, export_text
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, mean_squared_error, recall_score, precision_score
from sklearn.metrics import confusion_matrix
```

```python
modeling_df = pd.read_csv('./data/modeling_df.csv', index_col=False)
```

```python
modeling_df.sample(3)
```

# Preprocessing

### Define X and y:

```python
X = modeling_df['title']
y = modeling_df['is_il']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Define baseline accuracy
y.value_counts(normalize=True)
```

### Logistic Regression - TFIDFVectorizer Explore & Visualize

```python
# Transform X_train with TFIDFVectorizer
tvec = TfidfVectorizer(stop_words='english', ngram_range=(1,2), max_features=500)
Z_train = tvec.fit_transform(X_train)

# Instantiate and fit model
lr = LogisticRegression(solver='lbfgs', max_iter=5000)
lr.fit(Z_train, y_train)

# Score model on train data
lr.score(Z_train, y_train)
```

```python
# Train data cross val score
cross_val_score(lr, Z_train, y_train, cv=5).mean()
```

```python
# Transform X_test with TFIDFVectorizer
Z_test = tvec.transform(X_test)

# Test data score
lr.score(Z_test, y_test)
```

As with the rest of the supervised models explored, the logistic regression slightly overfit.

```python
# Create dataframe with all corpus words and text frequency inverse document frequency score
Z_train_df = pd.DataFrame(data=Z_train.toarray(),
                          columns=tvec.get_feature_names_out())
```

```python
# Look at first 3 rows of dataframe
Z_train_df.head(3)
```

```python
# Look at 20 words that appear most often across some documents, but not all documents
# AKA likely best predictors of subreddit classification
Z_train_df.sum().sort_values(ascending=False).head(20)
```

```python
# Look at strength of each of top 10 of the 500 features in predicting either subreddit
# Negative correlations --> Jewish ; Positive correlations --> Israel
features_500 = sorted(list(zip(lr.coef_[0], tvec.get_feature_names_out())))
features_500[:10] + features_500[-10:]
```

```python
# Make a dataframe of top 20 strongest predictors for Jewish
jew_feats_list = []

for x,y in features_500[:20]:
    feat = {}
    feat['feature'] = y
    feat['coefficient'] = x
    jew_feats_list.append(feat)

jew_feats_df = pd.DataFrame(jew_feats_list)
jew_feats_df.head()
```

```python
jew_feats_df.shape
```

```python
# Make a dataframe of top 20 strongest predictors for Israel
israel_feats_list = []

for x,y in features_500[:-21:-1]:
    feat = {}
    feat['feature'] = y
    feat['coefficient'] = x
    israel_feats_list.append(feat)

israel_feats_df = pd.DataFrame(israel_feats_list)
israel_feats_df.head()
```

```python
israel_feats_df.shape
```

```python
# Plot the top predictors for Jewish
plt.figure(figsize=(20,12))
sns.barplot(data=jew_feats_df, x='coefficient', y='feature').invert_xaxis()
plt.xlabel('Coefficient Strength', size=28)
plt.ylabel('Word', size=28)
plt.yticks(size=28)
plt.xticks(size=28)
plt.title('Top 20 Words with Strongest Predictive Power for\nJewish Subreddit', size=40)
sns.despine(left=True, bottom=True);
plt.savefig('./img/Top 20 Words with Strongest Predictive Power for Jewish Subreddit.png')
```

```python
# Plot the top predictors for Beatles
plt.figure(figsize=(20,12))
sns.barplot(israel_feats_df['coefficient'], israel_feats_df['feature'])
plt.xlabel('Coefficient Strength', size=28)
plt.ylabel('Word', size=28)
plt.yticks(size=28)
plt.xticks(size=28)
plt.title('Top 20 Words with Strongest Predictive Power for\nIsrael Subreddit', size=40)
sns.despine(left=True, bottom=True);
plt.savefig('./img/Top 20 Words with Strongest Predictive Power for Israel Subreddit.png')
```

```python
# Make a dataframe of middle predictors for both
jew_feats_list = []

for x,y in features_500[240:260]:
    feat = {}
    feat['feature'] = y
    feat['coefficient'] = x
    jew_feats_list.append(feat)

jewish_feats_df = pd.DataFrame(jew_feats_list)
jewish_feats_df.head()

# Plot the middle predictors for both
plt.figure(figsize=(20,12))
sns.barplot(data=jewish_feats_df, x='coefficient', y='feature').invert_xaxis()
plt.xlabel('Coefficient Strength', size=28)
plt.ylabel('Word', size=28)
plt.yticks(size=28)
plt.xticks(size=28)
plt.title('Next Top 20 Words with Strongest Predictive Power for\nJewish Subreddit', size=40)
sns.despine(left=True, bottom=True);
plt.savefig('./img/Next Top 20 Words with Strongest Predictive Power for Jewish Subreddit.png')
```

```python
# Get probability for 'subreddit' == 'Jewish'
probs = lr.predict_proba(Z_test)[:, 0]
probs[:10]
```

```python
# Get predictions
preds = lr.predict(Z_test)
preds[:10]
```

```python
# Zip probs / preds / actuals
probs_preds_actual = list(zip(probs, preds, y_test))
probs_preds_actual[:10]
```

```python
# Generate confusion matrix
cm = confusion_matrix(y_test, preds)
```

```python
# Create labeled dataframe from confusion matrix
cm_df = pd.DataFrame(cm, 
             columns=['Predicted Jewish', 'Predicted Israel'], 
             index=['Actual Jewish', 'Actual Israel'])
cm_df
```
